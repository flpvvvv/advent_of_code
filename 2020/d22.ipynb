```python
with open("d22_input.txt", "r") as f:
    lines = f.read().split("\n")
# lines = [(i[0], int(i[1:])) for i in lines]
lines[0], lines[-1]
```

```python
player_1 = []
player_2 = []
for idx, l in enumerate(lines):
    if l == "":
        break
    if l[0] == "P":
        continue
    player_1.append(int(l))
for i in range(idx + 1, len(lines)):
    l = lines[i]
    if l[0] == "P":
        continue
    player_2.append(int(l))

print(f"player_1: {player_1}")
print(f"player_2: {player_2}")
```

```python
class combat_game:
    def __init__(self, p1, p2):
        self.p1 = p1
        self.p2 = p2
        self.round = 0
        print(f"Player 1 has {len(self.p1)} cards.\nPlayer 2 has {len(self.p2)} cards.")

    def game_over(self):
        if len(self.p1) == 0:
            print("Player 2 wins!")
            return True
        if len(self.p2) == 0:
            print("Player 1 wins!")
            return True
        return False

    def play_one_round(self):
        if self.game_over():
            print("Game already over!!")
            return
        p1_card = self.p1[0]
        self.p1.remove(p1_card)
        p2_card = self.p2[0]
        self.p2.remove(p2_card)
        if p1_card > p2_card:
            self.p1.extend([p1_card, p2_card])
        else:
            self.p2.extend([p2_card, p1_card])
        self.round += 1
        # check if game over
        return self.game_over()

    def show_current_cards(self):
        print(f"After {self.round} rounds:")
        print(f"player_1: {self.p1}")
        print(f"player_2: {self.p2}")
```

```python
new_combat_game = combat_game(player_1, player_2)
```

```python
new_combat_game.show_current_cards()
new_combat_game.play_one_round()
new_combat_game.show_current_cards()
```

```python
player_1 = []
player_2 = []
for idx, l in enumerate(lines):
    if l == "":
        break
    if l[0] == "P":
        continue
    player_1.append(int(l))
for i in range(idx + 1, len(lines)):
    l = lines[i]
    if l[0] == "P":
        continue
    player_2.append(int(l))

print(f"player_1: {player_1}")
print(f"player_2: {player_2}")
```

```python
class recursive_combat_game:
    def __init__(self, p1, p2):
        self.p1 = p1
        self.p2 = p2
        self.round = 0
        self.all_rounds = []
        self.winner = 0
        # print(f"Player 1 has {len(self.p1)} cards.\nPlayer 2 has {len(self.p2)} cards.")

    def round_string(self):
        return (
            ",".join([str(i) for i in self.p1])
            + ":"
            + ",".join([str(i) for i in self.p2])
        )

    def alread_played(self):
        if self.round_string() in self.all_rounds:
            return True
        return False

    def game_over(self):
        if len(self.p1) == 0:
            self.winner = 2
            return True
        if len(self.p2) == 0:
            self.winner = 1
            return True
        return False

    def play_one_round(self):
        if self.winner != 0:
            return True
        p1_card = self.p1[0]
        self.p1.remove(p1_card)
        p2_card = self.p2[0]
        self.p2.remove(p2_card)
        if (len(self.p1) >= p1_card) and (len(self.p2) >= p2_card):
            # start a sub game
            sub_game = recursive_combat_game(self.p1[:p1_card], self.p2[:p2_card])
            while not sub_game.play_one_round():
                pass
            win_player = sub_game.who_wins()
            if win_player == 1:
                self.p1.extend([p1_card, p2_card])
            else:
                self.p2.extend([p2_card, p1_card])
        else:
            if p1_card > p2_card:
                self.p1.extend([p1_card, p2_card])
            else:
                self.p2.extend([p2_card, p1_card])
        if self.alread_played():
            self.winner = 1
            return True
        else:
            self.all_rounds.append(self.round_string())
        self.round += 1
        # check if game over
        return self.game_over()

    def show_current_cards(self):
        print(f"After {self.round} rounds:")
        print(f"player_1: {self.p1}")
        print(f"player_2: {self.p2}")

    def who_wins(self):
        if self.winner != 0:
            return self.winner
        else:
            if len(self.p1) == 0:
                self.winner = 2
                return 2
            if len(self.p2) == 0:
                self.winner = 1
                return 1
```

```python
new_recursive_combat_game = recursive_combat_game(player_1, player_2)
```

```python
while not new_recursive_combat_game.play_one_round():
    pass
new_recursive_combat_game.show_current_cards()
```

```python
winner_cards = (
    new_recursive_combat_game.p1
    if new_recursive_combat_game.who_wins() == 1
    else new_recursive_combat_game.p2
)
print(winner_cards)
```
