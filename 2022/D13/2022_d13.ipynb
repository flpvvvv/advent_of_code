```python
fp = "input.txt"
with open(fp, "r") as f:
    lines = f.read().split("\n")


def compare(l_li, r_li):
    l_value = l_li.pop(0) if len(l_li) > 0 else None
    r_value = r_li.pop(0) if len(r_li) > 0 else None

    if (l_value is None) and (r_value is not None):
        return 1
    if (l_value is not None) and (r_value is None):
        return -1
    if (l_value is None) and (r_value is None):
        return 0

    if isinstance(l_value, int) and isinstance(r_value, int):
        if l_value < r_value:
            return 1
        elif l_value > r_value:
            return -1
        else:
            return compare(l_li.copy(), r_li.copy())
    else:
        res = compare(
            l_value.copy() if isinstance(l_value, list) else [l_value],
            r_value.copy() if isinstance(r_value, list) else [r_value],
        )
        if res == 0:
            return compare(l_li.copy(), r_li.copy())
        else:
            return res


compare_res = [
    compare(json.loads(lines[idx * 3]), json.loads(lines[idx * 3 + 1]))
    for idx in range((len(lines) + 1) // 3)
]

print(f"Answer to Q1: {(np.where(np.array(compare_res)==1)[0].flatten()+1).sum()}")

org_li = [line for line in lines if len(line) > 0]
org_li.extend(["[[2]]", "[[6]]"])

from functools import cmp_to_key

org_li.sort(
    key=cmp_to_key(lambda x, y: compare(json.loads(x), json.loads(y))), reverse=True
)

print(f"Answer to Q2: {(org_li.index('[[2]]')+1)*(org_li.index('[[6]]')+1)}")
```
