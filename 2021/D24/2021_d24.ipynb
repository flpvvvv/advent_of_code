```python
%%time

import copy
from collections import defaultdict

from rich import print
from tqdm import tqdm

init_z = 0
final_z = 0

# 3 lists from quiz input
alpha_li = [1, 1, 1, 26, 1, 26, 1, 1, 1, 26, 26, 26, 26, 26]
beta_li = [14, 15, 15, -6, 14, -4, 15, 15, 11, 0, 0, -3, -9, -9]
gamma_li = [1, 7, 13, 10, 0, 13, 11, 6, 1, 7, 11, 14, 4, 10]

# Understand from quiz input
def calculate_z(pre_z, alpha, beta, gamma, w):
    z = pre_z if alpha == 1 else int(pre_z / 26)
    if beta > 10:
        x = 1
    else:
        x = 0 if (pre_z % 26 + beta) == w else 1
    return z * (25 * x + 1) + (w + gamma) * x


z_li = [[init_z]]

for i, (alpha, beta, gamma) in tqdm(
    enumerate(zip(alpha_li, beta_li, gamma_li)),
    total=len(alpha_li) - 1,
    desc="Get all POSSIBLE z values",
):
    if i == 13:  # Skip the last one
        break
    new_z = []
    for w in range(1, 10):
        for pre_z in z_li[i]:
            new_z.append(calculate_z(pre_z, alpha, beta, gamma, w))
    z_li.append(list(set(new_z)))

# Reverse z list
z_li = z_li[::-1]

valid_z_li = [[final_z]]

for i, (alpha, beta, gamma) in tqdm(
    enumerate(zip(alpha_li[::-1], beta_li[::-1], gamma_li[::-1])),
    total=len(alpha_li) - 1,
    desc="Get all VALID z values",
):
    if i == 13:  # Skip the last one
        break
    new_valid_z = []
    valid_z = valid_z_li[i]
    for pre_z in z_li[i]:
        for w in range(1, 10):
            z = calculate_z(pre_z, alpha, beta, gamma, w)
            if z in valid_z:
                new_valid_z.append(pre_z)
                break
    valid_z_li.append(new_valid_z)

# Reverse valid z list
valid_z_li = valid_z_li[::-1]

pre_z_dict = defaultdict(list)
pre_z_dict[init_z] = [0]

for i, (alpha, beta, gamma) in tqdm(
    enumerate(zip(alpha_li, beta_li, gamma_li)),
    total=len(alpha_li),
    desc="Get all VALID model number",
):
    z_dict = defaultdict(list)
    for w in range(1, 10):
        for pre_z in pre_z_dict.keys():
            z = calculate_z(pre_z, alpha, beta, gamma, w)
            if z in valid_z_li[i]:
                z_dict[z].extend([int(str(num) + str(w)) for num in pre_z_dict[pre_z]])

    pre_z_dict = copy.deepcopy(z_dict)

print(f"Anwer to Q1: {max(z_dict[final_z])}")
print(f"Anwer to Q2: {min(z_dict[final_z])}")
```
